### Importing Libraries 

```python
from mtcnn import MTCNN
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import img_to_array
from sklearn.preprocessing import LabelBinarizer,LabelEncoder
from tensorflow.keras.utils import to_categorical
import numpy as np
import cv2 as cv
import glob
import matplotlib.pyplot as plt
```

### Face Detection function

```python
detector_obj= MTCNN()

def face_detect(img):
    rgb_img = cv.cvtColor(img , cv.COLOR_BGR2RGB)

    face = detector_obj.detect_faces(rgb_img)[0]
    x , y , w , h = face['box']

    return rgb_img[y:y+h , x:x+w]# This crop only section of image that contain person Face
```

## Iterate Files:

```python
all_images = []
all_labels = []

for i , file in enumerate(glob.glob('smile_dataset/**/**')) :
    img = cv.imread(file)
    try :
        face = face_detect(img)
        face = cv.resize(face , (32,32))
        face = face/255.0
        
        all_images.append(face)
        
        #cv.imshow('myWin' , face)
        #if cv.waitKey(0) == ord('q') :
        #break
        
        label = file.split('\\')[-2]
        all_labels.append(label)
    except :
        pass
    
cv.destroyAllWindows()
```

## Encoding Labels


```python
lb = LabelBinarizer()
all_labels = lb.fit_transform(all_labels)

le = LabelEncoder()
all_labels_le = le.fit_transform(all_labels)

all_labels_le = to_categorical(all_labels_le)
```

### Spliting Data

```python
all_images = np.array(all_images)

X_train , X_test , y_train , y_test = train_test_split(all_images , all_labels_le , test_size=0.2 ,random_state=120)

print("X_train : ", X_train.shape)
print("X_test : ", X_test.shape)
print("y_train : ", y_train.shape)
print("y_test : ", y_test.shape)
```

### Creating Model

```python
model = models.Sequential([
    layers.Conv2D(32, (3,3) ,activation ='relu' , input_shape=(32, 32, 3)),
    layers.Conv2D(32, (3,3) ,activation ='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D(2,2),
    
    layers.Conv2D(64, (3,3) ,activation ='relu'),
    layers.Conv2D(64, (3,3) ,activation ='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),
    layers.Dense(32,activation ='relu'),
    layers.Dense(2,activation ='softmax')
])

model.compile(optimizer='sgd' ,loss = "categorical_crossentropy", metrics =['accuracy'])
```

### Fitting Model

```python
H = model.fit(X_train ,y_train , validation_data=(X_test,y_test) , epochs=10, batch_size=32)
```

### Plotting Train and Validation Accuracy

```python
plt.plot(H.history['accuracy'] , label='Train Accuracy')
plt.plot(H.history['val_accuracy'] , label='Test Accuracy')

plt.plot(H.history['loss'] , label='Train loss')
plt.plot(H.history['val_loss'] , label='Test loss')

plt.xlabel('Epoch')
plt.ylabel('Accuracy/Loss')
plt.legend()
plt.show()
```
